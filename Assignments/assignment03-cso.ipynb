## A program that retrieves the dataset for the "exchequer account (historical series)" from the CSO, and stores it into a file called "cso.json"

### URL: https://ws.cso.ie/public/api.jsonrpc?data=%7B%22jsonrpc%22:%222.0%22,%22method%22:%22PxStat.Data.Cube_API.ReadDataset%22,%22params%22:%7B%22class%22:%22query%22,%22id%22:%5B%5D,%22dimension%22:%7B%7D,%22extension%22:%7B%22pivot%22:null,%22codes%22:false,%22language%22:%7B%22code%22:%22en%22%7D,%22format%22:%7B%22type%22:%22JSON-stat%22,%22version%22:%222.0%22%7D,%22matrix%22:%22FIQ02%22%7D,%22version%22:%222.0%22%7D%7D

```python
import requests
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"
```

```python
payload = {
    "jsonrpc": "2.0",
    "method": "PxStat.Data.Cube_API.ReadDataset",
    "params": {
        "class": "query",
        "id": [],
        "dimension": {},        
        "extension": {
            "pivot": None,
            "codes": False,
            "language": {"code": "en"},
            "format": {"type": "JSON-stat", "version": "2.0"}
        },
        "matrix": "FIQ02"
    }
}
```

This section creates Python dictionary called `payload`. It reflects JSOn body that is copid from the CSO dataset page API tab.<br>
This is needed for the the below `request.post()` to work and convert to JSON.

```python
response = requests.post(url, json=payload)
# Process the response as needed - sends HTTP request to CSO API with the specified payload and stores the response in the 'response' variable.
```

```python
data = response.json()  # Converts the JSON response into a Python dictionary.

with open("cso.json", "w") as f:  # Opens a new file for writing using Python's built-in I/O system.
        f.write(response.text) # Writes the raw JSON text returned by the API into cso.json.
```
